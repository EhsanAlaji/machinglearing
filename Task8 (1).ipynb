```python
import numpy as np #math
import pandas as pd #files
import matplotlib.pyplot as plt #grahp
```

```python

path=r'C:\machine-learing\data\smoke_detection.txt'
data = pd.read_csv(path)
```

```python

yesfire =data[data['Fire_Alarm'].isin([1])]
nofire = data[data['Fire_Alarm'].isin([0])]
print(yesfire)
print('-------------')
print(nofire)
```

```python

data = [len(yesfire), len(nofire)] 
labels=['yes fire','no fire']
plt.figure(figsize=(8,10))
plt.pie(data, labels=labels, autopct="%1.1f%%")
plt.title('fire alarm graph')
plt.legend(title='fire alarm')
plt.show()
```
